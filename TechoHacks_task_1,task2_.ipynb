## Task-1  **Visualization using Histogram**

---



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
iris = pd.read_csv(url, names=column_names)
print(iris.head())

```

```python
sns.histplot(data=iris, x="sepal_length", kde=True)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=iris, x="sepal_width", kde=True)
plt.title("Histogram of Sepal width")
plt.xlabel("Sepal width")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=iris, x="petal_length", kde=True)
plt.title("Histogram of petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=iris, x="petal_width", kde=True)
plt.title("Histogram of petal width")
plt.xlabel("Petal width")
plt.ylabel("Frequency")
plt.show()
```

**scatterplot**

```python
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.show()
```
